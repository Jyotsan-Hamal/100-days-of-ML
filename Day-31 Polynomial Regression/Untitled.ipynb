```python
import numpy as np

import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression,SGDRegressor

from sklearn.preprocessing import PolynomialFeatures,StandardScaler

from sklearn.metrics import r2_score

from sklearn.pipeline import Pipeline
```

```python
x = 6 * np.random.randn(200,1)-3
y = 0.8 * x**2 + 0.9 * x + 2 + np.random.randn(200,1)

#y = 0.8x^2 + 0.9x + 2
```

```python
plt.plot(x,y,'b.')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
lr = LinearRegression()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
r2_score(y_test,y_pred)
```

```python

plt.scatter(x,y)
plt.plot(x_train,lr.predict(x_train),color='r')
```

```python
# Applying polynomial features
```

```python
ploy = PolynomialFeatures(degree=2)
x_train_tra = ploy.fit_transform(x_train)
x_tets_tra = ploy.fit_transform(x_test)
```

```python
print(x_train[0])
print(x_train_tra[0])
```

```python
Polynomial = LinearRegression()
Polynomial.fit(x_train_tra,y_train)
```

```python
Y = Polynomial.predict(x_tets_tra)
r2_score(y_test,Y)
```

```python
x_new = np.linspace(-3,3,200).reshape(200,1)
x_new_poly = ploy.transform(x_new)
y_new = Polynomial.predict(x_new_poly)
```

```python
plt.plot(x_new,y_new,'r-',color='r',linewidth=2,label='prediction')
plt.plot(x_test,y_test,'b.',color='b',label ='Training')
plt.plot(x_test,y_test,'g.',label='Testing Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
