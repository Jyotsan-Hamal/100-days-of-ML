```python
class Gradient_Descent:
    def __init__(self,learning_rate,epochs):
        self.slope = 35.12
        self.bias = -120
        self.lr = learning_rate
        self.epochs = epochs
    def fit(self,x,y):
        #calculate b using gradient descent
        for i in range(self.epochs):
            b_slope = -2*np.sum(y-self.slope*x.ravel()-self.bias)
            self.bias = self.bias - (self.lr*b_slope)
            print(self.bias,b_slope)
        print(self.bias)
        
```

```python
gd = Gradient_Descent(0.001,50)
```

```python
gd.fit(X,Y)
```
